```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Thanasis\Downloads\UNZIP_FOR_NOTEBOOKS_FINAL\DATA\Advertising.csv")
X = df.drop('sales',axis=1)
y = df['sales']
```

```python
df.head()
```

```python
#Regularization it wouldn't be so effective on such a small feature set. So we do regularization on the polynomial feature set
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polynomial_converter = PolynomialFeatures(degree=3,include_bias=False)
```

```python
poly_features = polynomial_converter.fit_transform(X)
```

```python
poly_features.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge_model = Ridge(alpha=10)
```

```python
ridge_model.fit(X_train,y_train)
```

```python
test_predictions = ridge_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
MAE = mean_absolute_error(y_test,test_predictions)
MSE = mean_squared_error(y_test,test_predictions)
RMSE = np.sqrt(MSE)
```

```python
MAE
```

```python
MSE
```

```python
RMSE
```

```python
# Training Set Performance (to be used later for comparison)
train_predictions = ridge_model.predict(X_train)
MAE = mean_absolute_error(y_train,train_predictions)
MAE
```

```python
#RidgeCV allows us to check multiple Alpha Hyperparameters in one run, together with Cross Validation

from sklearn.linear_model import RidgeCV
```

```python
# All score metric in Scikit-learn follow convention "Higher is better", so thats why negative mean absolute error
# See all options: sklearn.metrics.SCORERS.keys()

#In Parameter alphas we put in the hyperparameter we want to check for tuning the model with the best of em

#The default Cross Validation is Leave-One-Out. Since our data is small, we go with the default

ridge_cv_model = RidgeCV(alphas=(0.1, 1.0, 10.0),scoring='neg_mean_absolute_error')
```

```python
ridge_cv_model.fit(X_train,y_train)
```

```python
ridge_cv_model.alpha_
```

```python
test_predictions = ridge_cv_model.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test,test_predictions)
MSE = mean_squared_error(y_test,test_predictions)
RMSE = np.sqrt(MSE)
```

```python
# Training Set Performance
train_predictions = ridge_cv_model.predict(X_train)
MAE = mean_absolute_error(y_train,train_predictions)
MAE
```
